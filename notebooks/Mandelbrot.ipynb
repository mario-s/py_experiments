# Mandelbrot Set

marks the set of points in the complex plane

Mandelbrot set is the set obtained from the quadratic recurrence equation

https://mathworld.wolfram.com/MandelbrotSet.html

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

find if a complex number diverges under max_steps and if it does then when

```python
def get_iter(c:complex, thresh:int = 4, max_steps:int = 25) -> int:
    # Z_(n) = (Z_(n-1))^2 + c
    # Z_(0) = c
    z=c
    i=1
    while i < max_steps and (z*z.conjugate()).real < thresh:
        z=z*z +c
        i+=1
    return i
```

```python
def plotter(n, thresh, max_steps = 25):
    mx = 2.48 / (n-1)
    my = 2.26 / (n-1)
    
    #maps the image coordinate to the real-valued number in the range [-2,0.47] for x
    #and in the range [-1.12,1.12] for y 
    mapper = lambda x,y: (mx*x - 2, my*y - 1.13)
    
    #Numpy array of shape N*N, to store the pixel values of Mandelbrot Set
    img=np.full((n,n), 255)
    
    #loop through all pixels coordinates, using two loops, and calculate the pixel value for that coordinate
    for x in range(n):
        for y in range(n):
            it = get_iter(complex(*mapper(x,y)), thresh=thresh, max_steps=max_steps)
            img[y][x] = 255 - it
    return img
```

let's make some images

```python
img = plotter(n=1000, thresh=4, max_steps=50)
plt.imshow(img, cmap="plasma")
plt.axis("off")
plt.show()
```

```python
img = plotter(n=4000, thresh=4, max_steps=50)
plt.imshow(img, cmap="magma")
plt.axis("off")
plt.show()
```
